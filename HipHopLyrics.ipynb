The following cells are web scrapers that will retrieve the song titles and artists names from three webpages.

```python
#This code is to retrieve song titles from the website
response=urllib.request.urlopen('https://www.xxlmag.com/news/2016/12/50-hip-hop-best-songs-of-2016/')
#2017: https://www.xxlmag.com/news/2017/12/best-hip-hop-songs-2017/
#2018: https://www.xxlmag.com/news/2018/12/best-hip-hop-songs-2018/
html = str(response.read())
#regular expression code is important to pull title from html
r = re.compile('(?<=h2 data-reactid=\"\d\d\d\"\>).*?(?=\<\/h2)')
Song=r.findall(str(html))
```

```python
#This code is to retrieve artists from the website
response=urllib.request.urlopen('https://www.xxlmag.com/news/2016/12/50-hip-hop-best-songs-of-2016/')
#2017: https://www.xxlmag.com/news/2017/12/best-hip-hop-songs-2017/
#2018: https://www.xxlmag.com/news/2018/12/best-hip-hop-songs-2018/
html = str(response.read())
#regular expression code is important to pull artist name from html
r = re.compile('(?<=small data-reactid=\"\d\d\d\"\>).*?(?=\<\/small)')
Artist=r.findall(str(html))
```

The next 10 cells will clean the data within the dataframe. This step is needed so the next webscraper can accurately
retrieve lyrics based on the data points.

```python
#Cleaning the song title to use in the spider to grasp lyrics
B=list(df['Song'])
B1=[]
for b in B:
    b=b.replace('"',' ')
    b=b.strip()
    b=b.replace(" ","")
    b=b.lower()
    B1.append(b)
df['lowerSong']=B1
```

```python
#Remove non-alphanumeric characters from song titles; this will be used for lyrics spider
lowerSong=[]
for i in range(len(df['lowerSong'])):
    song=df['lowerSong'][i]
    for j in song:
        if j.isalpha() or j.isnumeric():
            next
        else:
            song=song.replace(j,'')
    lowerSong.append(song)       

#Replaced column with clean values for song titles    
df['lowerSong']=lowerSong
```

```python
#Cleaning the artist name to use in the spider to grasp lyrics
A=list(df['Artist'])
Search=[]
L=[]
for a in A:
    #obtaining the main artist for each song
    a=a.replace(" ","")
    i=a.find('Featuring')
    #find the index where featuring begins then split at that point
    if i>=0:
        a=a[0:i]
    j=a.find('and')
    if j>=0:
        a=a[0:j]
    Search.append(a.lower())

#To ensure the artist name is only alpha characters; Otherwise it needs more cleaning
#The names that need more cleaning is in list L
for s in Search:    
    if s.isalpha():
        next
    else:
        L.append(s)
```

The next cell will obtain the lyrics from AZ lyrics and add the column to the dataframe. The cell afterwards will clean the lyrics to my liking.

```python
#This is the web scraper created to obtain lyrics for all the songs
#I used similar python libraries to the Spider Deliverable as well as html_text to retrieve the text in the webpage
Lyrics=[]
for i in range(len(df)):
    try:
        response=urllib.request.urlopen('https://www.azlyrics.com/lyrics/'+df['lowerArtist'][i]+'/'+df['lowerSong'][i]+'.html')
        html = str(response.read())
        Yes=ht.extract_text(html)
        Lyrics.append(Yes)
        time.sleep(5)
    except:
        Lyrics.append('Not Available')
        next
        time.sleep(5)

df['Lyrics']=Lyrics
```

```python
#I used this code to clean up the text (lyrics) from the html pages. I recognize that this was not the best method 
#but it works for the scope of my project. In the future, I will try other python libraries.
def cleanLyrics(df,column):
    List=list(df[column])
    NewList=[]
    for i in range(len(List)):
        if df[column][i] == 'Not Available':
            NewList.append('Not Available')
        else:
            song=df[column][i]
            k=song.find('Search')
            j=song.find('Submit')
            song=song[k:j]
            for j in song:
                if j.isalpha() or j.isnumeric():
                    next
                else:
                    song=song.replace(j,' ')
            song=song.replace('r n','')
            song=song.replace('                                                                           ','')
            song=song.lower()
            NewList.append(song)
    return NewList
            #return c
            
Please=cleanLyrics(df,'lyrics')  
df['lyrics']=Please
```
